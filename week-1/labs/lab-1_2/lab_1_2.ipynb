### q5. Type casting
Convert "15" (string) → int and add to a number.  
Convert "3.5" → float and multiply.  
Convert 99 → string and concatenate.  
Convert different values to bool and observe truthiness.

```python
# Convert "15" (string) → int and add to a number.
a = "15"
b = 20
print(int(a) + b)

# Convert "3.5" → float and multiply.
c = "3.5"
print(b * float(c))

# Convert 99 → string and concatenate.
d = 99
print(str(d) + " Your marks is this bro you are amazing.")

# Convert different values to bool and observe truthiness.
# 0 and 1
print(f"bool(0): {bool(0)}")
print(f"bool(1): {bool(1)}")
print(f"bool(-3): {bool(-3)}")

# with strings (empty is False, non-empty is True)
print(bool(""))   # False
print(bool(" "))   # True
print(bool("hello"))   # True

# # Collections (empty is False, non-empty is True)
print(bool([])) # False
print(f"bool([1,2]): {bool([1,2])}") # True
print(f"bool({{}}): {bool({})}")   # False
print(f"bool(()): {bool(())}")   # False
print(f"bool((1,2)): {bool((1,2))}")   # True

# Special values
print(f"bool(None): {bool(None)}")     # False')
print(f"bool(3.14): {bool(3.14)}")
```
